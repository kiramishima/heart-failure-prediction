# Visualizations

```python
# Group by Sex
grp_by_sex = df.group_by("Sex").agg(total=pl.len()).sort(by="total", descending=True)

fig, ax = plt.subplots()

g = sns.barplot(grp_by_sex, x='Sex', y='total', ax=ax)
g.set(
    title="Total Men & Woman",
    ylabel="Total"
)

plt.show()
```

```python
# Histograms and KDE

numeric_values = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak', 'HeartDisease']

fig, ax = plt.subplots(2, 3, figsize=(18, 9))
for idx, column in enumerate(numeric_values[:3]):
    sns.histplot(data=df[column], ax=ax[0][idx-1], kde=True)
    ax[0][idx-1].set_title(column)
    
for idx, column in enumerate(numeric_values[3:]):
    sns.histplot(data=df[column], ax=ax[1][idx-1], kde=True)
    ax[1][idx-1].set_title(column)

plt.show()
```

```python
# Boxplots

fig, ax = plt.subplots(2, 3, figsize=(18, 9))
for idx, column in enumerate(numeric_values[:3]):
    sns.boxplot(data=df[column], ax=ax[0][idx-1])
    ax[0][idx-1].set_title(column)
    
for idx, column in enumerate(numeric_values[3:]):
    sns.boxplot(data=df[column], ax=ax[1][idx-1])
    ax[1][idx-1].set_title(column)

# Show the plot
plt.tight_layout()
plt.show()
```
